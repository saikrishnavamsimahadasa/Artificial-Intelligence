```python
def plot_sensitivity(xs, ys, yerr, name, xlabel,ylabel):
    %matplotlib notebook
    plt.figure()
    plt.errorbar(xs,ys, yerr=yerr,marker='o',linestyle='None',capsize=5,c='mediumslateblue')
    #plt.legend()
    plt.title("Sensitivity analysis: {}".format(name))
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.savefig("figures/{}.png".format(name.replace(" ","_")))
```

```python
#Binomial test px > py
#x: # wins for agent x
#y: # wins for agent y
#n: # runs for agent x
#m: # runs for agent y
#z_alpha: critical value for 95% that result is significant
def binomial_test(x,y,n=20,m=20, z_alpha=1.64):
    p_hat = 1.0*(x+y)/(n+m)
    px = 1.0*x/n
    py = 1.0*y/m
    z = 1.0*(px-py)/(p_hat*(1-p_hat)/n + p_hat*(1-p_hat)/m)**0.5
    reject_h0 = z > z_alpha
    if reject_h0:
        print "Can conclude px > py, critical value {}".format(z)
    else:
        print "Cannot conclude px > py, critical value {}".format(z)

print "MC > AB, small adversarial"
binomial_test(16,9)
print "MC > Exp, small adversarial"
binomial_test(16,6)
print "MC > AB, small adversarial"
binomial_test(18,9)
print "MC > AB, small adversarial"
binomial_test(18,10)
```
